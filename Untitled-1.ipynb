```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.feature_extraction.text import CountVectorizer # CountVectorizer for bag of words model
from sklearn.neural_network import MLPClassifier # Multilayer Perceptron Classifier for classification
import string 
from textblob import TextBlob # TextBlob for sentiment analysis
```

```python
data = pd.read_csv("SMSCollection.csv", encoding='latin-1') # read csv file
# map ham to 0 and spam to 1
data['Label'] = data['Label'].map({'ham': 0, 'spam': 1}) # map ham to 0 and spam to 1

# remove extra spaces
data['Message'] = data['Message'].str.replace('\s+', ' ')


# apply 5 features to the data
data['length'] = data['Message'].apply(len) # length of the message
data['capital_letters'] = data['Message'].apply(lambda x: sum(1 for c in x if c.isupper())) # number of capital letters
data['exclamation_marks'] = data['Message'].apply(lambda x: sum(1 for c in x if c == '!')) # number of exclamation marks
data['punctuation_count'] = data['Message'].apply(lambda x: sum(1 for c in x if c in string.punctuation)) # number of punctuation marks
data['sentiment_score'] = data['Message'].apply(lambda x: TextBlob(x).sentiment.polarity) # sentiment score
```

```python
# transform the Message column to a 1D array of floating points
vectorizer = CountVectorizer(analyzer = "word",   \
                             tokenizer = None,    \
                             preprocessor = None, \
                             stop_words = None,   \
                             max_features = 5000)       # create a CountVectorizer object with the parameters above

data_features = vectorizer.fit_transform(data['Message']) # fit the vectorizer to the training data
data_features = data_features.toarray() # convert to numpy array
```

```python
# using supervised algorithm train this neural network by splitting test data and training data
from sklearn.model_selection import train_test_split # import train_test_split for splitting the data into training and test data

X_train, X_test, y_train, y_test = train_test_split(data_features, data['Label'], test_size=0.2, random_state=0) # split the data into training and testing sets
```

```python
# training

# create a MLPClassifier object and train it
import pickle # import pickle for saving the model
filename = 'finalized_model.sav' # name of the file to save the model

train=False # set to True to train the model


if train:                                                                   # if train is True
    num_iterations = 3                                                      # number of iterations to train the model
    for i in range(num_iterations):                                         # train the model for 3 iterations
        mlp = MLPClassifier(hidden_layer_sizes=(100,50), solver='lbfgs', alpha=1e-5, random_state=i) # create a MLPClassifier object with the parameters above
        mlp.fit(X_train, y_train)                                           # fit the model to the training data
        print("Training iteration: " + str(i))                              # print the iteration number
        print("Training Accuracy: " + str(mlp.score(X_train, y_train)))     # print the training accuracy
        print("Testing Accuracy: " + str(mlp.score(X_test, y_test)))        # print the testing accuracy
        print("\n")                                                         # print a new line
    pickle.dump(mlp, open(filename, 'wb'))                                  # save the model

else:                                                                       # if the model is already trained
    mlp = pickle.load(open(filename, 'rb'))                                 # load the model

    


```

```python

# predict function that will take one single message and add 5 features to it and feed it to model and output the label
# check if model exists on disk

def predict(model,message):

    data = pd.DataFrame()
    data['Message'] = [message]
    data['length'] = data['Message'].apply(len)
    data['capital_letters'] = data['Message'].apply(lambda x: sum(1 for c in x if c.isupper()))
    data['exclamation_marks'] = data['Message'].apply(lambda x: sum(1 for c in x if c == '!'))
    data['punctuation_count'] = data['Message'].apply(lambda x: sum(1 for c in x if c in string.punctuation))
    data['sentiment_score'] = data['Message'].apply(lambda x: TextBlob(x).sentiment.polarity)

    
    data['Message'] = data['Message'].apply(lambda x: x.strip())


    message_feature = vectorizer.transform(data['Message'])
    # transform the other features to a 1D array of floating points

    message_feature = message_feature.toarray()
    
    label = model.predict(message_feature)
    
    if label == 0:
        print("Not Spam")
        return "Not Spam"
    else:
        print("Spam")
        return "Spam"

import os

# get user input
message = input("Enter your message: ")
if os.path.isfile(filename):
    # load model from disk
    loaded_model = pickle.load(open(filename, 'rb'))
    predict(loaded_model,message)
else:
    predict(mlp,message)
```
